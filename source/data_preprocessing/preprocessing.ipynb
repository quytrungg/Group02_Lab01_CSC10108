## Import dữ liệu cần thiết

```python
import pandas as pd 
```

## Bước 1: Đọc file CSV

```python
covid_df = pd.read_csv('../data/original/March_06_2023_corona_data.csv')
covid_df
```

## Bước 2: Xem và chuyển đổi kiểu dữ liệu

```python
covid_df.dtypes
```

```python
covid_df['Continent'] = covid_df['Continent'].fillna('Unknown')

covid_df.head(10)
```

## Bước 3: Xóa hết các dấu "+" có trong dữ liệu của 2 thuộc tính "New Cases" và "New Recovered"

```python
covid_df['New Cases'] = covid_df['New Cases'].apply(lambda x: x.replace('+', '') 
                                                    if isinstance(x, str) 
                                                    else x)
covid_df['New Recovered'] = covid_df['New Recovered'].apply(lambda x: x.replace('+', '') 
                                                            if isinstance(x, str) 
                                                            else x)

covid_df.head(10)
```

## Bước 4: Convert các cột có kiểu dữ liệu số về lại đúng kiểu dữ liệu của nó:

Các cột sau đây đã được đưa về về đúng kiểu dữ liệu rồi nên bỏ qua:
- 1 Test every X ppl     
- New Cases/1M pop       
- New Deaths/1M pop      
- New Deaths 

```python
numeric_cols = ['Total Cases', 'New Cases', 'Total Deaths', 'Total Recovered',
                'New Recovered', 'Active Cases', 'Serious, Critical', 'Tot Cases/1M pop',
                'Deaths/1M pop', 'Total tests', 'Tests/1M pop', 'Population',
                '1 Case every X ppl', '1 Death every X ppl', 'Active Cases/1M pop']

covid_df[numeric_cols] = covid_df[numeric_cols].apply(lambda x: x.str.replace(',', '')).astype(float)
covid_df = covid_df.fillna(0)
```

## Bước 5: Drop 2 quốc gia "Diamond Princess" và "MS Zaandam":

2 quốc gia "Diamond Princess" và "MS Zaandam" thiếu rất nhiều dữ liệu, không giúp được gì cho quá trình quan sát các mối quan hệ nên nhóm quyết định drop 2 dòng này.

```python
to_drop = covid_df[covid_df["Country"].isin(["Diamond Princess", "MS Zaandam"])].index
covid_df = covid_df.drop(to_drop)
```

## Bước 6: Trả về file CSV.

```python
covid_df.to_csv('../data/modified/Modified_March_06_2023_corona_data.csv', index=False) 
```
